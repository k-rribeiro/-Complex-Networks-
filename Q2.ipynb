**Questão 1** - Solicitado que calculemos media da distancia do menor caminho e o diametro, utilizando o arquivo 'hamster.txt' entregue

```python
#bibliotecas utilizadas
import numpy as np
#biblioteca usada all_pairsa criar e maniplar grafos
import networkx as nx
#biblioteca all_pairsa plotar caso necessario
import matplotlib.pyplot as plt
```

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('hamsterster.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

#faz o grafo
Grafo = nx.Graph()
Grafo.add_edges_from(ver)

#encontrando o maior componente conectado
MaiorCC = max(nx.connected_components(Grafo), key=len)

#aqui criamos um sub grafo onde ha esse maior componente
Grafo_conect = Grafo.subgraph(MaiorCC)

#calculamos aqui a media do caminho mais curto
cmc = nx.average_shortest_path_length(Grafo_conect)
print("Média da distancia do menor caminho: ", cmc)

#calcula o diametro do grafo
diametro = nx.diameter(Grafo_conect)
print('Network diameter:', diametro)
```

**Questão 2** - Solicitado que calculemos a media e a variancia do menor caminho, utilizando o arquivo 'USairport500.txt' entregue

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('USairport500.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

#monta o grafo
Grafo = nx.Graph()
Grafo.add_edges_from(ver)

#encontrando o maior componente conectado
MaiorCC = max(nx.connected_components(Grafo), key=len)

#aqui criamos um sub grafo onde ha esse maior componente
Grafo_conect = Grafo.subgraph(MaiorCC)

#calculamos aqui a media do caminho mais curto
cmc = nx.average_shortest_path_length(Grafo_conect)
print("Média da distancia do menor caminho: ", cmc)

# aqui iremos calcular o comprimento do caminho mais curto no subgrafo conectado
todosOSmenoresCaminhos = dict(nx.all_pairs_shortest_path_length(Grafo_conect))

#cria um array para colocar as distancias
distances = []
#armazena o comprimento do caminho mais curto na lista
for source, paths in todosOSmenoresCaminhos.items():
    for target, length in paths.items():
        if source != target:
            distances.append(length)

# aqui fica o calculo da variancia
variancia = np.var(distances, ddof=0)
print(f"Variância:", variancia)
```

**Questão 4** - na questão é solicitado que calculemos a entropia de shannon levando em consideração o tamanho dos menores caminhos, utilizando o arquivo 'USairport500.txt' entregue

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('USairport500.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

#monta o grafo
Grafo = nx.Graph()
Grafo.add_edges_from(ver)

#funcao para calcular a entropia de shannon com os nos
def entropia_shannon_caminhos(Grafo):

    #aqui criamos uma lista para armazenar o tamanho de todos os menores caminhos
    menorcamin = []

    # calcula os menores caminhos
    for node in Grafo.nodes():
        try:
            tam = nx.shortest_path_length(Grafo, source=node)
            menorcamin.extend([d for d in tam.values() if d > 0])
        except nx.NetworkXNoPath:
            continue

    # caso não tenhamos caminhos válidos vai retornar 0
    if not menorcamin:
        return 0

    # calcula aqui a distribuição de probabilidade dos comprimentos
    lengths_u, counts = np.unique(menorcamin, return_counts=True)
    prob = counts / counts.sum()
    #evitar log 0
    prob = prob[prob > 0]

    #calcula de fato a entropia
    entropia = -np.sum(prob * np.log2(prob))

    return round(entropia, 7)


Q3 = entropia_shannon_caminhos(Grafo)
print("Entropia de Shannon:", Q3)
```

**Questão 5** - Solicita que calculemos o coeficiente de correlacao de pearson entre os graus medios dos vizinhos e o grau da rede, utilizando o arquivo 'word_adjacencies.txt' entregue

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('word_adjacencies.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

#monta o grafo
Grafo = nx.Graph()
Grafo.add_edges_from(ver)
Grafo.remove_edges_from(nx.selfloop_edges(Grafo))

#seleciona o maior componente
Gcc = sorted(nx.connected_components(Grafo), key=len, reverse=True)
Grafo = Grafo.subgraph(Gcc[0])

#converte o rótulo dos nós em inteiros para facilitar a ordenação
G = nx.convert_node_labels_to_integers(Grafo, first_label=0)

#aqui vamos guardar o numero de nos e o numero de arestas neste grafo
N = len(Grafo)
M = G.number_of_edges()

#aqui montamos uma matriz com todas as distancias entre os pares de nós
if nx.is_connected(Grafo) == True:
    #D é a matriz
    D = np.zeros(shape=(N,N))
    vl = []
    for i in np.arange(0,N):
        for j in np.arange(i+1, N):
            if(i != j):
                aux = nx.shortest_path(Grafo,i,j)
                dij = len(aux)-1
                D[i][j] = dij
                D[j][i] = dij
                vl.append(dij)


#grau médio de uma vizinhança de uma rede
knn = []

for i in Grafo.nodes():
    aux =  nx.average_neighbor_degree(Grafo, nodes = [i])
    knn.append(float(aux[i]))
knn = np.array(knn)

#armazenamos o grau de cada nó presente no grafo em uma lista
vk = dict(Grafo.degree())
vk = list(vk.values())

#calcula a correlação entre a conectividade de uma vizinhança e o grau na rede
knnk = list()
ks = list()
for k in np.arange(np.min(vk), np.max(vk)+1):
    aux = vk == k
    if(len(knn[aux]) > 0):
        av_knn = np.mean(knn[aux])
        knnk.append(av_knn)
        ks.append(k)

#chama a função e depois imprime
coefPEARSON = np.corrcoef(ks, knnk)[0,1]
print('coeficiente de correlaçao de Pearson:', coefPEARSON)
```
