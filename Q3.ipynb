**Questão 1** - Solicita que calcule a media da medida de eigenvector centrality da rede USairport500

```python
#bibliotecas utilizadas
import numpy as np
#biblioteca usada all_pairsa criar e maniplar grafos
import networkx as nx
#biblioteca all_pairsa plotar caso necessario
import matplotlib.pyplot as plt
#importa do scipy as funcoes prontas da correlacao de pearson e de spearman
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('USairport500.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))


# Cria o grafo
Grafo = nx.Graph()
Grafo.add_edges_from(ver)


#transforma o grafo para não direcionado
Grafo = Grafo.to_undirected()
#tira as arestas onde o no está conectado a si mesmo
Grafo.remove_edges_from(nx.selfloop_edges(Grafo))
#aqui ele acha todos os componentes conectados do grafo e os organiza do maior para o menor e por fim os armazena
Gcc = sorted(nx.connected_components(Grafo), key=len, reverse=True)
#aqui deixa so o maior componente conectado do grafo e descarta todos os outros componentes menores
Grafo = Grafo.subgraph(Gcc[0])
#aqui renomeia os nos para inteiros sequenciais começando em 0, assim padronizando-os
Grafo = nx.convert_node_labels_to_integers(Grafo, first_label=0)

#calcula a centralidade do autovetor
centrality = nx.eigenvector_centrality(Grafo)
#exyrai os valores e os converte para uma lista  edessa lista vira um array onde podemos fazer operações
eigencentra = (list(centrality.values()))
#calcula a media dos valores da centralidade
media_eigenvector = np.mean(eigencentra)


print('Media da medida de eigenvector centrality', media_eigenvector)
```

**Questão 2** - Pede para que calculemos a correlação de pearson entre as medida betweenness centrality e grau para a rede solicitada

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('hamsterster.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

# Cria o grafo
Grafo = nx.Graph()
Grafo.add_edges_from(ver)


#transforma o grafo para não direcionado
Grafo = Grafo.to_undirected()
#tira as arestas onde o no está conectado a si mesmo
Grafo.remove_edges_from(nx.selfloop_edges(Grafo))
#aqui ele acha todos os componentes conectados do grafo e os organiza do maior para o menor e por fim os armazena
Gcc = sorted(nx.connected_components(Grafo), key=len, reverse=True)
#aqui deixa so o maior componente conectado do grafo e descarta todos os outros componentes menores
Grafo = Grafo.subgraph(Gcc[0])
#aqui renomeia os nos para inteiros sequenciais começando em 0, assim padronizando-os
Grafo = nx.convert_node_labels_to_integers(Grafo, first_label=0)

#calcula a Betweenness Centrality
betweenness = nx.betweenness_centrality(Grafo)
grau = dict(Grafo.degree())

#aqui extrai os nos e os valores da betweenness
nos = list(Grafo.nodes())
#armazena os valores de betweenness centrality para cada no do grafo
b_valor = np.array([betweenness[n] for n in nos])
grauvalor = np.array([grau[n] for n in nos])

#calcula a correlação de Pearson a partir de uma funcao pronta da biblioteca scipy
corr, p_valor = pearsonr(b_valor, grauvalor)


print("correlação de pearson entre betweenness centrality e o grau:", corr)
```
