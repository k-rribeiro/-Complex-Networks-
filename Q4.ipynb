```python
#bibliotecas utilizadas
import numpy as np
#biblioteca usada para criar e manipolar grafos
import networkx as nx
#biblioteca para plotar caso necessario
import matplotlib.pyplot as plt
from networkx.algorithms.community import greedy_modularity_communities
import community as community_louvain
```

```python
#metodo para adquirir as informações dadas em arquivo txt earmazenar numa lista
with open('jazz.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

G = nx.read_edgelist("jazz.txt", nodetype=int)
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))


Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0]).copy()
G = nx.convert_node_labels_to_integers(G, first_label=0)

communities = greedy_modularity_communities(G)

mod = nx.algorithms.community.quality.modularity(G, communities)
print(f"Modularidade (FastGreedy): {mod:.4f}")
```

**Questão 3 -**

```python
mu_values = [0.05, 0.1, 0.2]
N = 128
tau1 = 3
tau2 = 1.5
k = 16
minc = 32
maxc = 32

for mu in mu_values:
    try:
        #gera o grafo lrf
        G = nx.LFR_benchmark_graph(
            n=N, tau1=tau1, tau2=tau2, mu=mu,
            min_degree=k, max_degree=k,
            min_community=minc, max_community=maxc,
            seed=10
        )

        G = G.to_undirected()
        G.remove_edges_from(nx.selfloop_edges(G))

        #aqui vamos verificar se o grafo não está vazio
        if len(G) == 0:
            print(f"μ = {mu:.2f} → Grafo vazio gerado")
            continue

        partition = community_louvain.best_partition(G)
        modularity = community_louvain.modularity(partition, G)

        print(f"μ = {mu:.2f} e modularidade: {modularity:.4f}")

    except Exception as e:
        print(f"Erro")
```
