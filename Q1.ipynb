**Questão 1** - é solicitado que a partir das informações dadas pelo arquivo 'lesmis.txt' calcule o terceiro e quarto momento do grafo criado



```python
#bibliotecas utilizadas

import numpy as np
#biblioteca usada para criar e maniplar grafos
import networkx as nx
#biblioteca para plotar caso necessario
import matplotlib.pyplot as plt
```

```python
#Para realizar esse questionário, estou utilizando o Google Colab. Nele, podemos colocar os arquivos de dados entregue e usados
#na mesma pasta do algoritmo do Colab, tornando mais fácil sua leitura.

#o comando lê o arquivo txt linha por linha, armazenando-o na variavel "armazena"
with open('lesmis.txt', 'r') as file:
  armazena = file.readlines()

#primeiro cria-se de uma lista vazia para armazenar as arestas e pesos que serão pegos linha por linha
ver = []
for line in armazena:
    #nesta parte remove qualquer espaço em branco que tenha entre os elementos da lista
    source, target, weight = map(int, line.strip().split())
    ver.append((source, target, weight))

#cria o grafo vazio
Grafo = nx.Graph()

#colocando as informações que temos dentro desse grafo criado, como a orientação das arestas e seus pesos
for edge in ver:
    source, target, weight = edge
    Grafo.add_edge(source, target, weight=weight)

#função que calcula a distribuição do grau
def distriGRAU(Grafo):

    #uma lista com os graus de todos os nós do grafo
    grau = [Grafo.degree(node) for node in Grafo.nodes()]
    #descobre qual o grau maximo do grafico
    grauMAX = max(grau)
    # uma lista de contagem de tamanho grauMAX+1 para incluir todos os graus de 0 até maximo
    noqq = [0] * (grauMAX + 1)

    for k in grau:
        noqq[k] += 1
    #verifica e armazena o numero total de nós presente no grafo
    totalnos = len(grau)
    #armazena uma tal probabilidade de nó um qualquer ter grau m
    Pdek = [count / totalnos for count in noqq]
    #uma lista que representa todos os valores possiveis para o grau
    valorespo = list(range(grauMAX + 1))
    #retorna a lista das probabilidades
    return valorespo, Pdek

#função que vai calcular o momento m do grau(com base no solicitado pela questão)
def momento(Grafo, mom):

    k, Pdek = distriGRAU(Grafo)
    #realiza o somatorio da sequência dos momentos
    M = np.sum(np.fromiter(((k_i**mom) * p_i for k_i, p_i in zip(k, Pdek)), dtype=float))
    #retorna o momento m desejado
    return M


#chamada da função para o terceiro e o quarto momento
terceiroMomento = momento(Grafo, 3)
quartoMomento = momento(Grafo, 4)

print("Terceiro momento:", terceiroMomento)
print("Quarto momento:", quartoMomento)
```

**Questão 2** - com as informações dadas no arquivo 'powergrid.txt' é solicitado que a calcule o valor da complexidade do grafo


```python
#utilizando o mesmo comando para conseguir ler e armazenas os dados entregues
with open('powergrid.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

#cria um novo grafo, porém diferente do primeiro ele não é ponderado
Grafo = nx.Graph()
#grafo não possui pesos, então armazenamos apenas suas arestas
Grafo.add_edges_from(ver)

#função que calcula o coeficiente de complexidade
def coefcomplex(Grafo):
    #para calcular esse coeficiente precisamos do primeiro e segundo momento
    mo1 = momento(Grafo, 1)
    mo2 = momento(Grafo, 2)

    #variavel responsavel por armazenar e calcular o coeficiente
    alfa = mo2 / mo1
    return alfa

print("Coeficiente de complexidade:", coefcomplex(Grafo))
```

**Questão 3** - Com as informações dadas no arquivo 'euroroad.txt' é solicitado que calcule a Entropia de Shannon do grafo

```python
#utilizando o mesmo comando para conseguir ler e armazenas os dados entregues
with open('euroroad.txt', 'r') as file:
    armazena = file.readlines()

ver = []
for line in armazena:
    source, target = map(int, line.strip().split())
    ver.append((source, target))

# cria um novo grafo não ponderado
Grafo = nx.Graph()
# grafo não possui pesos, então armazenamos apenas suas arestas
Grafo.add_edges_from(ver)

#Nesta função calculamos a entropia de Shannon
def entropia_shannon(Grafo):

    graus = np.asarray([d for _, d in Grafo.degree()])
    graus_u, counts = np.unique(graus, return_counts=True)
    prob = counts / counts.sum()
    prob = prob[prob > 0]
    #PK*log_2 Pk
    entropia = -np.sum(prob * np.log2(prob))
    return round(entropia, 7)

print("Entropia da distribuição de graus:", entropia_shannon(Grafo))
```
